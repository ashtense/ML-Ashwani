```python
# These imports will enable us to access the data directory and read csv files.
import pandas as pd
import numpy as np
import os
```

```python
# All the dependencies needed for building the model
import keras
# Importing the sequential model. There is another model class used with the functional API.
from keras.models import Sequential
# Importing the layers of our model. As we have decided upon using dense layers. Hence here we go.  
from keras.layers import Dense
```

```python
# First lets prepare the data from concrete data set as discussed in detail in the notebook:
#DL0101EN-3-1-Regression-with-Keras-py-v1.0__2_.ipynb
data_file = os.path.join(os.path.dirname(os.getcwd()), 'datasets', 'concrete_data.csv')
concrete_data = pd.read_csv(data_file)
concrete_data.head()
concrete_data.shape
concrete_data.describe()
concrete_data.isnull().sum()
```

```python
# Lets check the shape of the dataset
concrete_data.shape
# Prints: (x, y) where x is the number of rows and y is the number of columns.
# There seems to be around 1000 rows of data to train the model.
```

```python
# Lets check the data for any missing values
concrete_data.describe()
concrete_data.isnull().sum()
# If all the columns null count sum is 0, then it means the data is clean for the model to be trained
```

```python
# Lets split the training data into predictors and targets as needed for training with forward propagation.
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
predictors.head()
```

```python
# Normalization of the data set: subtracting the mean and dividing the remaining by standard deviation
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
# Lets define the regression model inside a function so that we can call it later as per our requirement.
def regression_model(input_nodes):
    model = Sequential()
    # Add the first layer i.e. the input layer, number of columns as per the training data set.
    model.add(Dense(5, activation='relu', input_shape=(input_nodes,)))
    # Adding the first hidden layer
    model.add(Dense(5, activation='relu'))
    # Adding the output layer
    model.add(Dense(1))
    
    # Compiling the model
    """ For training purposes we need to define an optimizer and an error metric
    We used gradient descent as our minimization and optimization algorithm
    We used the mean squared error as our error measurement between the predicted value and ground truth
    adam algorithm is more efficient then gradient descent
    adam doesn't require the learning rate as was needed for gradient descent algorithm """
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python

# Number of columns in the prediction dataset of concrete data
n_cols = predictors_norm.shape[1]
# Lets initialize the model
model = regression_model(n_cols)
```

```python
# Next, we will train and test the model at the same time using the *fit* method. 
# We will leave out 30% of the data for validation and we will train the model for 100 epochs.
model.fit(predictors, target, validation_split=0.3, epochs=1000, verbose=1)
```

```python
# Once the training is done using predictors and target datasets, we can use the model to predict the output for new data
test_data = pd.DataFrame([401.8 ,94.7 ,0.0 ,147.4 ,11.4 ,946.8 ,852.1 ,28]).transpose()
predictions = model.predict(test_data)
print(predictions)
```
